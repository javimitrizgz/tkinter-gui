```python
from tkinter import *
from tkinter import ttk

class StartScreen:
    def open_new_window(self):
        self.master.withdraw()
        self.new_wind_root=Toplevel(self.master)
#         pass this as master to the root level window
        new_wind=AdminMainScreen(self.new_wind_root)
    
    def __init__(self, master):
#         store the master refernce as a class variable
        self.master=master
        self.label=ttk.Label(self.master,
                             text="Catalogued")
        self.label.config(font=("Courier",30))
        self.label.grid(row=0,column=0,columnspan=2)
        # username related
        labelUsername=Label(self.master, 
                            text="Username",
                            fg="SeaGreen3")
        labelUsername.config(font=("Courier", 20))
        entryUsername =Entry(self.master, 
                             bd =1,
                             width=30)
        labelUsername.grid(row=5,column=0,pady=20,padx=20)
        entryUsername.grid(row=5,column=1,pady=20)
        # password related
        labelPass=Label(self.master,
                        text="Password",
                        fg="SeaGreen3")
        labelPass.config(font=("Courier", 20))
        entryPass = Entry(self.master,
                          show='*',
                          bd =1,
                          width=30)
        labelPass.grid(row=8,column=0, pady=20, padx=20)
        entryPass.grid(row=8,column=1)
        # login button
        btnContinue=Button(self.master,
                           text="Continue",
                           height=1, width=20,
                           command=lambda: self.open_new_window())
        btnContinue.grid(row=9,column=1,columnspan=2)
#         btnContinue.bind('<Button-1>',open_new_window)
    
class AdminMainScreen:
    
#     constructor for AdminMainScreen
    def __init__(self,master):
#         sore reference to master as a class variable
        self.master=master
    
        self.label=ttk.Label(self.master,text="Catalogued")
        self.label.config(font=("Courier",30))
        self.label.grid(row=0,column=1,columnspan=2)
#         create new model button
        self.btnCreateModel=Button(self.master,
                                   text="Create New Model",
                                   height=1, width=20,
                                   command=lambda: self.open_create_new_model_screen()) #launch create new model screen
        self.btnCreateModel.grid(row=1,column=1,columnspan=2)
#         add update model button
        self.btnUpdateModel=Button(self.master,
                                   text="Update Model",
                                   height=1, width=20,
                                   command=lambda: self.open_update_model_screen())
        self.btnUpdateModel.grid(row=2,column=1,columnspan=2)
        self.btnCreateModel.grid(row=3,column=1,columnspan=2)
        
#         create delete model button
        self.btnCreateModel=Button(self.master,
                                   text="Delete Model",
                                   height=1, width=20,
                                   command=lambda: self.open_delete_model_screen())
#         add update model button
        self.btnUpdateModel=Button(self.master,
                                   text="Rename Model",
                                   height=1, width=20,
                                   command=lambda: self.open_rename_model_screen())
        self.btnUpdateModel.grid(row=4,column=1,columnspan=2)
    
#         add Generate video button
        self.btnUpdateModel=Button(master,
                                   text="Generate Video",
                                   height=1, width=20,
                                   command=lambda: self.open_generate_video_screen()) #launch generate videoo screen
        self.btnUpdateModel.grid(row=5,column=1,columnspan=2)
        
    def open_create_new_model_screen(self):
        self.master.withdraw()
        self.root_for_create_new_model_sceen=Toplevel(self.master)
#         open the window by providing it the above created ref as its master
        create_vid_screen=CreateNewModelScreen(self.root_for_create_new_model_sceen)
    
    def open_update_model_screen(self):
        self.master.withdraw()
        self.root_for_update_model_window=Toplevel(self.master)
#         open the window by providing it the above created ref as its master
        update_window=CreateNewModelScreen(self.root_for_update_model_window)
        
    def open_delete_model_screen(self):
        self.master.withdraw()
        self.root_for_delete_model_window=Toplevel(self.master)
        update_window=CreateNewModelScreen(self.root_for_delete_model_window)
        
    def open_rename_model_screen(self):
        self.master.withdraw()
        self.root_for_rename_model_window=Toplevel(self.master)
        update_window=RenameModelScreen(self.root_for_rename_model_window)

    def open_generate_video_screen(self):
        self.master.withdraw()
        self.root_for_generate_vid_screen=Toplevel(self.master)
        gen_vid_screen=GenerateVideoScreen(self.root_for_generate_vid_screen)
    
    
class CreateNewModelScreen:
    
    def __init__(self,master):
        
        self.master=master
        
         # model name related
        self.label=ttk.Label(self.master,
                             text="Model Name")
        self.label.config(font=("Courier",15))
        self.label.grid(row=0,column=0,columnspan=1)
        entrySubject =Entry(self.master,
                            bd =1,
                            width=30)
        entrySubject.grid(row=0,column=1)
        
#         add create model button
        self.btnCreate=Button(self.master,
                                   text="Create Model",
                                   height=1, width=20,
                                   command=lambda: print('creating new model'))
        self.btnCreate.grid(row=1,column=1,columnspan=2)
        
        
class UpdateModelScreen:
    
    def __init__(self,master):
#         store reference to master as class var
        self.master=master
        
#        model name related
        self.label=ttk.Label(self.master,
                             text="Model Name")
        self.label.config(font=("Courier",15))
        self.label.grid(row=0,column=0,columnspan=1)
        entrySubject =Entry(self.master,
                            bd =1,
                            width=30)
        entrySubject.grid(row=0,column=1)
        
#         add update model button
        self.btnUpdateModel=Button(self.master,
                                   text="Update Model",
                                   height=1, width=20,
                                   command=lambda: print('updating the model'))
        self.btnUpdateModel.grid(row=1,column=1,columnspan=2)
        
        
class DeleteModelScreen:
    def __init__(self,master):
        
        self.master=master
        
#         model name related
        self.label=ttk.Label(self.master,
                             text="Model Name")
        self.label.config(font=("Courier",15))
        self.label.grid(row=0,column=0,columnspan=1)
        entrySubject =Entry(self.master,
                            bd =1,
                            width=30)
        entrySubject.grid(row=0,column=1)
        
#         add delete the model button
        self.btnDelete=Button(self.master,
                                   text="Delete the Model",
                                   height=1, width=20,
                                   command=lambda: print('Deleting the chosen model'))
        self.btnDelete.grid(row=1,column=1,columnspan=2)
        

class RenameModelScreen:
     def __init__(self,master):
#         store reference to master as class var
        self.master=master
        
#        model name related
        self.label=ttk.Label(self.master,
                             text="Model Name")
        self.label.config(font=("Courier",15))
        self.label.grid(row=0,column=0,columnspan=1)
        entrySubject =Entry(self.master,
                            bd =1,
                            width=30)
        entrySubject.grid(row=0,column=1)
        
#         new model name related
        self.label=ttk.Label(self.master,
                             text="New Model Name")
        self.label.config(font=("Courier",15))
        self.label.grid(row=1,column=0,columnspan=1)
        entrySubject =Entry(self.master,
                            bd =1,
                            width=30)
        entrySubject.grid(row=1,column=1)
        
#         add the rename model button
        self.btnDelete=Button(self.master,
                                   text="Rename the Model",
                                   height=1, width=20,
                                   command=lambda: print('Renaming the chosen model'))
        self.btnDelete.grid(row=2,column=1,columnspan=2)
        
class GenerateVideoScreen:
    def __init__(self,master):
        
        self.master=master
        
#         subject label related
        self.label=ttk.Label(self.master,
                             text="Subject Label")
        self.label.config(font=("Courier",15))
        self.label.grid(row=0,column=0,columnspan=1)
        entrySubject =Entry(self.master,
                            bd =1,
                            width=30)
        entrySubject.grid(row=0,column=1)
        
#         add Generate video button
        self.btnGenerateVideo=Button(self.master,
                                   text="Generate Subject's Footage",
                                   height=1, width=20,
                                   command=lambda: print('Generating subject\'s video'))
        self.btnGenerateVideo.grid(row=1,column=1,columnspan=2)
        

def main():
    root_window = Tk()
    root_window.geometry('400x300')
    root_window.resizable(width=False,height=False)
    app=StartScreen(root_window)
    root_window.mainloop()

if __name__=="__main__": main()
    
```
